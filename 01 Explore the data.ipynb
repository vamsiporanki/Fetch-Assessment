```python
engine='python'
import pandas as pd

# Load datasets
df_products = pd.read_csv("C:/Users/vamsi/Desktop/vizva con/Fetch Assessment/PRODUCTS_TAKEHOME.csv")
# Display each dataset
display(df_products)

```

```python
df_transactions = pd.read_csv("C:/Users/vamsi/Desktop/vizva con/Fetch Assessment/TRANSACTION_TAKEHOME.csv")
display(df_transactions)
```

```python
df_users = pd.read_csv("C:/Users/vamsi/Desktop/vizva con/Fetch Assessment/USER_TAKEHOME.csv")
display(df_users)
```

```python
#Missing values percentage
missing_products = df_products.isnull().mean() * 100
missing_transactions = df_transactions.isnull().mean() * 100
missing_users = df_users.isnull().mean() * 100

print("Missing values in PRODUCTS dataset:\n", missing_products)
print("Missing values in TRANSACTION dataset:\n", missing_transactions)
print("Missing values in USER dataset:\n", missing_users)
```

```python
#Duplicate values
duplicates_products = df_products.duplicated().sum()
duplicates_transactions = df_transactions.duplicated().sum()
duplicates_users = df_users.duplicated().sum()

print("Duplicate records in PRODUCTS:", duplicates_products)
print("Duplicate records in TRANSACTIONS:", duplicates_transactions)
print("Duplicate records in USERS:", duplicates_users)
```

```python
#Identify Data type issues
print(df_products.dtypes)
print(df_transactions.dtypes)
print(df_users.dtypes)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")

# Visualization 1: Missing values heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df_products.isnull())
plt.title("Missing Values in PRODUCTS Dataset")
plt.show()
```

```python
missing_values = pd.DataFrame({
    "Products": df_products.isnull().mean() * 100,
    "Transactions": df_transactions.isnull().mean() * 100,
    "Users": df_users.isnull().mean() * 100
}).reset_index().melt(id_vars="index", var_name="Dataset", value_name="Missing %")
```

```python
plt.figure(figsize=(14, 6))
sns.barplot(x="index", y="Missing %", hue="Dataset", data=missing_values)
plt.xticks(rotation=90)
plt.title("Percentage of Missing Values Per Column")
plt.show()
```
